```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#MÃ¡scaras.
sobel_x = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
sobel_y = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
roberts_xy = np.array([[1, 0], [0, -1]])
roberts_yx = np.array([[0, 1], [-1, 0]])
laplace = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
```

1) Aplique os filtros de realce Sobel e Roberts nas imagens abaixo. Mostre, para o caso do Sobel, uma imagem representando a derivada em x e outra imagem representando a derivada em y.

```python
#AplicaÃ§Ã£o do filtro Gaussiano para remoÃ§Ã£o de ruÃ­dos, o que facilita a identificaÃ§Ã£o das bordas.
img1 = cv2.GaussianBlur(cv2.imread("cln1.png", cv2.IMREAD_GRAYSCALE),(3,3),0)

#ConvoluÃ§Ã£o. 
img1_sobel_x = cv2.filter2D(img1, -1, sobel_x)
img1_sobel_y = cv2.filter2D(img1, -1, sobel_y)

img1_roberts_xy = cv2.filter2D(img1, -1, roberts_xy)
img1_roberts_yx = cv2.filter2D(img1, -1, roberts_yx)

fig = plt.figure(figsize=(15, 45))
rows = 1
columns = 4

fig.add_subplot(rows, columns, 1)
plt.imshow(img1_sobel_x, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Sobel X")

fig.add_subplot(rows, columns, 2)
plt.imshow(img1_sobel_y, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Sobel Y")

fig.add_subplot(rows, columns, 3)
plt.imshow(img1_roberts_xy, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Roberts XY")

fig.add_subplot(rows, columns, 4)
plt.imshow(img1_roberts_yx, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Roberts YX")
```

2) Utilizando a mesma imagem, realce as bordas utilizando (a) Filtro Laplaciano, (b) Unsharp masking, (c) Highboost filtering (escolher um k) e (d) K <1 para atenuar.

```python
img1_laplace = cv2.filter2D(img1, -1, laplace)

img1_um = img1 - img1_laplace
img1_hf_a = img1 - 2 * img1_laplace
img1_hf_b = img1 - 0.5 * img1_laplace

fig = plt.figure(figsize=(15, 45))
rows = 1
columns = 4

fig.add_subplot(rows, columns, 1)
plt.imshow(img1_laplace, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Laplace")

fig.add_subplot(rows, columns, 2)
plt.imshow(img1_um, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Unsharp Masking")

fig.add_subplot(rows, columns, 3)
plt.imshow(img1_hf_a, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Highboost Filtering (k = 2)")

fig.add_subplot(rows, columns, 4)
plt.imshow(img1_hf_b, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Highboost Filtering (k = 0.5)")
```
