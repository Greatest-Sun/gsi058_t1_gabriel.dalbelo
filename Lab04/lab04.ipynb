```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#A função blur da biblioteca OpenCV pode ser usada para aplicar o filtro de média em uma imagem.
#Kernel é uma pequena matrix quadrada (n x n), sendo k = n.

img1 = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)

#k = 3.
img1a = cv2.blur(img1,(3, 3))

#k = 3, aplicado 3x seguidas na mesma imagem.
img1aa = cv2.blur(img1a,(3, 3))
img1aaa = cv2.blur(img1aa,(3, 3))

#k = 7
img1b = cv2.blur(img1,(7, 7))

fig = plt.figure(figsize=(15, 45))
rows = 1
columns = 4

fig.add_subplot(rows, columns, 1)
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(img1a, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("3x3")

fig.add_subplot(rows, columns, 3)
plt.imshow(img1aaa, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("3x3 (3 vezes seguidas)")

fig.add_subplot(rows, columns, 4)
plt.imshow(img1b, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("7x7")
```

```python
img2 = cv2.imread("ben2.png", cv2.IMREAD_GRAYSCALE)

#GaussianBlur(imagem, (k, k), sigma)
img2a = cv2.GaussianBlur(img2,(5,5),1)
img2b = cv2.GaussianBlur(img2,(9,9),2)
img2c = cv2.GaussianBlur(img2,(15,15),4)

fig = plt.figure(figsize=(15, 45))
rows = 1
columns = 4

fig.add_subplot(rows, columns, 1)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(img2a, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Kernel = 5x5, Sigma = 1")

fig.add_subplot(rows, columns, 3)
plt.imshow(img2b, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Kernel = 9x9, Sigma = 2")

fig.add_subplot(rows, columns, 4)
plt.imshow(img2c, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Kernel = 15x15, Sigma = 4")
    
```

```python
#medianBlur(imagem, k)
img3 = cv2.imread("sta2noi1.png", cv2.IMREAD_GRAYSCALE)
img3a = cv2.medianBlur(img3, 3)
img3aa = cv2.medianBlur(img3a, 7)

fig = plt.figure(figsize=(15, 45))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img3, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(img3a, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("3x3")

fig.add_subplot(rows, columns, 3)
plt.imshow(img3aa, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title("3x3 -> 7x7")
```
